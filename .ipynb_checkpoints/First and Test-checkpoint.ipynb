Resampling will give take the multivariate time series data and resample them for a different time period. 
<a href="https://leve.rs/blog/data-analysis-using-python-pandas/">A nice description of how this works</a>
Aggregate only works with the group by column values. But resampling can find sum/avg/std of values spanning multiple days or just hours.

````python
weather = pd.read_csv('https://raw.githubusercontent.com/jvns/pandas-cookbook/master/data/weather_2012.csv', 
                     index_col='Date/Time', parse_dates=True, dayfirst=True)
print(weather.head())
print(weather['Temp (C)'][:5])
mean_temp = weather['Temp (C)'].resample('D', how='mean') # daily mean
print(mean_temp.head())
````

````python
data['mean temp'] = mean_temp
data.head()
````

````python
data[['Berri 1', 'mean temp']].plot(subplots=False)
data[['Berri 1', 'mean temp']].plot(subplots=True)
````

````python
import numpy as np
arr = np.array([1,2,3,4,5])
print(arr)
print(len(arr))
print(arr.dtype)
print(arr.astype(str))
print('-' * 20) ## Underline with string concatenation!

mat = np.array([ [1,2,3,4], [2,3,4,5], [3,4,5,6]])
print(mat.ndim) # Dimension
print(mat.shape)# m,n
print(mat.size) # num of elements
print(mat)

print(np.mean(mat)) # mean of all elements
print(np.mean(mat, axis=0)) # col-wise mean
print(np.mean(mat, axis=1)) # row-wise mean

print(mat[0])
print(mat[0:2]) # excluding row 2. 
print(mat[:2,:2]) # First 2 rows and first 2 columns
print(mat[0][::-1]) ## Take first row and reverse it!
````

We can use masked_array to ignore <b>nan or np.nan</b> values which would otherwise mess up the calculations

````python
invalid = np.array([[1.023, 2.045], [np.nan, 56.234]])
print(invalid)
print(invalid.mean()) # Problem
print(np.isnan(invalid)) # Mask
masked = np.ma.masked_array(invalid, np.isnan(invalid))
print(masked)
print(masked.mean())
````

## Importing files, images, plotting
**os - getcwd, path.join, listdir**<br/>
**matplotlib.pyplot - imread**<br />
You can use pyplot to read the image too. You can also use matplotlib.image to read the image using mpimg.imread command. 
Use the command 
```python %matplotlib inline
``` to show the image in the browser itself. Else a separate window will be opened. To show the image, plt.show should be called after plt.imshow

````python
%matplotlib inline
import matplotlib.pyplot as plt
#import matplotlib.image as mpimg
import os

cwd = os.getcwd()
image_dir = os.path.join(cwd, 'images')
files = [file for file in os.listdir(image_dir) if file.endswith('jpg')]

img = plt.imread(os.path.join(image_dir,'stinkbug.png'))
print(img.shape)
plt.imshow(img)
plt.show()
````

````python
# %matplotlib osx
redperimage =[]
greenperimage = []
np.seterr(divide='ignore', invalid='ignore') # to ignore divide by zero error which is 
                                             # properly handled by np.array

for file in files:
    if ('stink' in file):
        continue
    img = plt.imread(os.path.join(image_dir, file))
    print(file)
    reds = img[:,:,0]
    redperimage.append(np.sum(reds))
    greens = img[:,:,1]
    greenperimage.append(np.sum(greens))
#ratio = redperimage/greenperimage
print(redperimage)
rpi = np.array(redperimage, dtype=float)
gpi = np.array(greenperimage, dtype=float)
print(rpi)
print(gpi)
ratio = rpi / gpi
print(ratio)

plt.subplot(211)
plt.plot(range(0,len(redperimage)), redperimage, 'r*') # range(0,len(redperimage)) for x index 
                                                       # of files. r* - mark with red stars
plt.plot(range(0,len(greenperimage)), greenperimage, 'go') # go - green dots

plt.subplot(212)
plt.plot(range(0, len(ratio)), ratio, 'ko') # black dots
plt.title('Ratio of red to green')
````

````python
def expo(a):
    count = 0;
    for i in list(a):
        for j in list(a):
            count = int(i)+int(j)

expo_vect = np.vectorize(expo)
````

````python
%%timeit a="123417286815618398573292798379285"
expo(a)
````

````python
%%timeit a="123417286815618398573292798379285"
expo_vect(a)
````

Converting a string into a list of numbers? Use try except to handle invalid strings

````python
inputString = '10 12 14 15.6 0.34 inv 096.56 1000'
output = []
for i in inputString.split():
    try:
        num = eval(i)
    except:
        num = 0
    output.append(num)
output
````

If you use int(i) instead of eval(i), the floating point numbers will be invalid too.
