```python
%matplotlib inline
# To find the contribution variation across time,
fec.groupby(['party','contb_receipt_dt'])['contb_receipt_amt'].sum()
#But this would look aweful if you try to plot it because 
# contb_receipt_dt is string
```

```python
# Convert string date to numerical date
str_date = fec.contb_receipt_dt[0]
[day,month,year] = str_date.split('-')
print('-'.join([day,str(months[month]),year]))
```

```python
def date_convert(str_date):
    [day,month,year] = str_date.split('-')
    return '-'.join([day,str(months[month]),year])
print(fec.contb_receipt_dt.apply(date_convert).head())
```

```python
# Or better, convert it into datetime
def date_convert(str_date):
    day,month,year = str_date.split('-')
    return datetime(int('20'+year),months[month],int(day))
fec.contb_receipt_dt.map(date_convert)
```

```python
fec['contb_receipt_dt'] = fec.contb_receipt_dt.map(date_convert)
```

```python
by_date = fec.groupby(['party','contb_receipt_dt'])['contb_receipt_amt'].sum()
```

```python
by_date
```

```python
#As we can see above, now the date column is sorted
by_date.unstack('party').head()
```

```python
# So now, to track how the amount varies over time, we take the cumulative sum
by_date.unstack('party').cumsum().head()
```

```python
by_date.unstack('party').cumsum().plot()
```
