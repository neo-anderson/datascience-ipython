###K Means Clustering

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

x = [1,5,1.5,8,1,9]
y = [2,8,1.8,8,0.6,11]
plt.scatter(x,y)

#X = np.array([])
# Convert the individual lists to a 2D list
arr = [[x[i],y[i]] for i in range(len(x))]
print(arr)
X = np.array(arr)
print(X)
```

```python
clf = KMeans(n_clusters = 2)
clf.fit(X)
centroids = clf.cluster_centers_
labels = clf.labels_
colors = ['r.','g.']
for i in range(len(X)):
    plt.plot(X[i][0],X[i][1],colors[labels[i]],markersize=12)
plt.plot(centroids[:,0],centroids[:,1],'k*',markersize=12)
```

###Hierarchical Clustering with Mean Shift

```python
from sklearn.cluster import MeanShift  # only good upto 10,000 samples?
from sklearn.datasets.samples_generator import make_blobs

centers = [[1,1],[5,5],[3,10]]
X,_ = make_blobs(n_samples = 500, centers = centers, cluster_std = 1)
plt.scatter(X[:,0],X[:,1])
```

```python
ms = MeanShift()
ms.fit(X)
labels = ms.labels_  # Predicted labels
cluster_centers = ms.cluster_centers_
colors = ['r.','g.','b.']
for i in range(len(X)):
    plt.plot(X[i,0],X[i,1],colors[labels[i]])
plt.plot(cluster_centers[:,0],cluster_centers[:,1],'k*',markersize=12)
print('Estimated cluster centers :',cluster_centers)
```
